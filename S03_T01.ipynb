Crea al menos una visualización para:

- Una variable categórica (UniqueCarrier)
- Una variable numérica (ArrDelay)
- Una variable numérica y una categórica (ArrDelay y UniqueCarrier)
- Dos variables numéricas (ArrDelay y DepDelay)

- Más de tres variables (ArrDelay, DepDelay, AirTime y UniqueCarrier).


```python
#primero importamos las librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#importamos el dataset
airlines_delay = pd.read_csv('DelayedFlights.csv')
```

### Se crea una visualizacion para la variable categoria Unique Carrier

```python
var_cat_UC = sns.catplot(x="UniqueCarrier", y='FlightNum', kind="bar", data=airlines_delay)
```

### Se crea una visualizacion para una variable numerica mediante un histograma

```python
histograma_arr= sns.histplot(data=airlines_delay, x="ArrDelay", palette='blue') #Crea un histograma
plt.xlim(0, 400)
```

### Se crea una visualizacion para una variable numérica y una categórica (ArrDelay y UniqueCarrier) 

```python
cat_num_bar = sns.barplot(x=airlines_delay['UniqueCarrier'],y=airlines_delay['ArrDelay'],data=airlines_delay)
```

### Dos variables numéricas
Para el analisis de dos variables categoricas se selecciono un scatter plot, debido a que en la representacion visual resulta mas clara la relacion entre las dos variables.

```python
dos_num = sns.scatterplot(x=airlines_delay['DepDelay'],y=airlines_delay['ArrDelay'],data=airlines_delay)
```

### Dos variables numericas y una variable categorica.
En la visualizacion de las dos variables numericas se integra una variable categorica agrupando los datos de acuerdo a una escala de colores asociada a los elementos de la variable 'uniquecarrier'. 

```python
var_cat2 = sns.scatterplot(x='ArrDelay', y='DepDelay', data=airlines_delay,
                hue='UniqueCarrier',) #agrupa los datos por 'carrier'
var_cat2.legend(bbox_to_anchor=(1.05, 1), borderaxespad=0.) #cambia la ubicacion de la leyenda del grafico
```

### Más de tres variables (ArrDelay, DepDelay, AirTime y UniqueCarrier).
Dentro de la misma visualizacion generada, se agrega el 'Airtime' como parametro distintivo del tamaño de los puntos asociados a cada variable.

```python
more_var= sns.relplot(
    data=airlines_delay,
    x="ArrDelay", y="DepDelay",
    hue="UniqueCarrier", size="AirTime", sizes=(10, 200))
```
